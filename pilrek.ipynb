```python
import pandas as pd
```

```python
url = 'https://github.com/eueung/pilrek/raw/master/pilrek.csv'
df  = pd.read_csv(url)
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df.shape
df.dtypes
```

```python
df.isna().sum()
```

```python
df['CaRek Pilihan'].value_counts().plot(kind='bar')
```

```python
ax = df['Kategori Anda'].value_counts().plot(kind='pie',autopct='%1.1f%%')
ax.set_ylabel('')
```

```python
ax = df['Kategori Umur Anda'].value_counts().plot(kind='pie',autopct='%1.1f%%')
ax.set_ylabel('')
```

```python
ax = df['Alasan Memilih CaRek'].value_counts().plot(kind='bar')
ax.set_ylabel('')
```

```python
new = df.groupby("CaRek Pilihan").get_group("Dicky Rezady Munaf (FSRD)")
new['Kategori Anda'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
newx = df.groupby("Kategori Anda").get_group("Mahasiswa")
newx['CaRek Pilihan'].value_counts().plot(kind='bar')
```

```python
civitas = df.loc[(df['Kategori Anda'] == 'Mahasiswa') | (df['Kategori Anda'] == 'Dosen')]
civitas['CaRek Pilihan'].value_counts().plot(kind='bar').set_ylabel('Sebaran Pilihan Civitas Akademika')
```
